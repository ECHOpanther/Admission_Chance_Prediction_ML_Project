# Data Pre-Processing

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x)
print(y)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
print(x)
```

# Plotting Pairplots

```python
sns.pairplot(df)
```

# Backward Elimination

```python
import statsmodels.api as sm
def backwardElimination(x,sl):
    nums=len(x[0])
    for i in range(0, nums):
        stats = sm.OLS(endog=y,exog=x).fit()   
        maxVar = max(stats.pvalues).astype(float) 
        if maxVar > sl:
            for j in range(0, nums-i):
                if (stats.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1) 
                    print(stats.summary())
    print(stats.summary())
    return x                

x_temp=x[:,:]
sl=0.05
x_opt = backwardElimination(x_temp,sl)
print(len(x_opt[0]))
print(x_opt[0])
```

# Applying MLR Model after Backward-Elimination

```python
from sklearn.model_selection import train_test_split
x1_tr,x1_te,y1_tr,y1_te = train_test_split(x_opt,y,test_size=0.03,random_state=31) 

from sklearn.linear_model import LinearRegression
lr_opt = LinearRegression()
lr_opt.fit(x1_tr,y1_tr)

from sklearn.metrics import r2_score,mean_squared_error
y_pred1 = lr_opt.predict(x1_te)
r2=r2_score(y1_te,y_pred1)
mse = mean_squared_error(y1_te,y_pred1)
print(mse)  
print(r2) 

plt.scatter(x1_te,y1_te,c='red')
plt.scatter(x1_te,y_pred1,c='blue')
plt.show()
```

# Applying RFR Model after Backward-Elimination

```python
#from sklearn.model_selection import train_test_split
x2_tr,x2_te,y2_tr,y2_te = train_test_split(x_opt,y,test_size=0.03,random_state=31)

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=20,random_state=0)  
rfr.fit(x2_tr,y2_tr)

#from sklearn.metrics import r2_score,mean_squared_error
y_pred2=rfr.predict(x2_te)
r2=r2_score(y2_te,y_pred2)
mse = mean_squared_error(y2_te,y_pred2)
print(mse)  
print(r2)

plt.scatter(x2_te,y2_te,c='red')
plt.scatter(x2_te,y_pred2,c='blue')
plt.show()
```
